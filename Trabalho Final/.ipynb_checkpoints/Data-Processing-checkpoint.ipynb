```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import scale
```

```python
mnist_train = pd.read_csv('train.csv')
target = mnist_train[[0]].values.ravel()
train = mnist_train.iloc[:,1:].values
test = pd.read_csv("test.csv").values

```

```python
# standard deviation and mean vector of training data

a = np.matrix(train)
a = a.astype(float)

sigma = np.apply_along_axis(lambda x: np.std(x), axis=0, arr=a)
mu = np.apply_along_axis(lambda x: np.mean(x), axis=0, arr=a)

```

```python
import matplotlib.pyplot as plt

%matplotlib inline

pixel = range(0,784)
plt.plot(pixel, sigma, 'bo')
plt.show()
```

```python
# Cov Matrix

CoV = np.divide(sigma, mu)
CoV = np.nan_to_num(CoV)

plt.plot(pixel, CoV, 'bo')
plt.show()
```

```python
logCoV = np.log10(CoV)

plt.plot(pixel, logCoV, 'bo')
plt.show()
```

```python
from sklearn import decomposition

pca = decomposition.PCA()
a = scale(a)

pca.fit(a)
```

```python
plt.plot(pca.explained_variance_ratio_[0:100], linewidth=2)
plt.xlabel('Generated Feature')
plt.ylabel('Fraction of Variance')
plt.show()
```

```python
X_new = pca.transform(a)
```

```python
# SVM and PCA
```

```python
pca = decomposition.PCA(n_components=0.8, whiten=True)
train_x = pca.fit_transform(train)
test_x = pca.fit_transform(test)
```

```python
from sklearn import svm
```

```python
svc = svm.SVC(kernel='rbf', C=2)
svc.fit(train_x, target)
```

```python
test_y = svc.predict(test_x)
```

```python
pd.DataFrame({"ImageId": range(1,len(test_y)+1), "Label": test_y}).to_csv('out.csv', index=False, header=True)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import cross_validation

print "Preparing the data..."
print "--------------------------------"
print "Loading the train dataset..."
train = np.loadtxt("train.csv", delimiter=",", skiprows=1)
print "Loading the test dataset..."
test = np.loadtxt("test.csv", delimiter=",", skiprows=1)

print "separate labels from training data"
train_data = train[:,1:]
target = train[:,0]

#number of components to extract
print "creating PCA object..."
pca = PCA(n_components=40)

#fit the training data
print "fitting the training data..."
pca.fit(train_data)

#determine the amount of variance explained by components
print "calculating the amount of variance explained by components..."
sum_variance_ratio = np.sum(pca.explained_variance_ratio_)
print "amount of variance: {0}".format(sum_variance_ratio)
print "-------------------------------------------------"
print "About 79% of  the total variance in the dataset is explained by 40 components"


print "Extracting the features..."
train_ext = pca.fit_transform(train_data)
print "Shape of training data after dimensionality reduction: {0}".format(train_ext.shape)

print "Transform the test data using existing parameters..."
test_ext = pca.transform(test)
print "test data shape: {0}".format(test_ext.shape)

print "Normalizing the dataset to use nearest neighbors..."
min_max_scaler = MinMaxScaler()
train_norm = min_max_scaler.fit_transform(train_ext)
test_norm = min_max_scaler.fit_transform(test_ext)
```

```python
print "Training the data..."
print "-----------------------------"

def train_knn(n_neighbors, train_norm, target):
    knn = KNeighborsClassifier(n_neighbors)
    knn.fit(train_norm, target)
    return knn


print "Training with n_neighbors=5 ..."
knn = train_knn(5, train_norm, target)
print "evaluating performance with cross-validation..."
print cross_validation.cross_val_score(knn, train_norm, target, cv=5)

print "------------------------------------"

print "Training with n_neighbors=3 ..."
knn = train_knn(3, train_norm, target)
print "evaluating performance with cross-validation..."
print cross_validation.cross_val_score(knn, train_norm, target, cv=5)
```

```python
pred = knn.predict(test_norm)
```

```python
# write to a file
save = pred.round()
ind = np.arange(1,len(pred) + 1)
new_save = np.column_stack((ind, save))
np.savetxt('knn3pca.csv',new_save,delimiter=',',fmt='%0.0f')
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn import svm, grid_search
```

```python
print "Preparing the data..."
print "--------------------------------"

print "Loading the train dataset..."
train_data = np.loadtxt("train_without_target_and_border.csv", delimiter=",")
print "Loading the target dataset..."
target = np.loadtxt("target.csv", delimiter=",")
print "Loading the test dataset..."
test_data = np.loadtxt("test_without_border.csv", delimiter=",")


train_data_normalized = preprocessing.normalize(train_data, norm='l2')
test_data_normalized = preprocessing.normalize(test_data, norm='l2')
#number of components to extract
print "Reduction ..."
pca = PCA(n_components=35, whiten=True)
print "target shape: {0}, train shape: {1}".format(target.shape, train_data_normalized.shape)
print target

pca.fit(train_data_normalized)


print "transform training data..."
train_data_normalized = pca.transform(train_data_normalized)
print "transform test data..."
test_data_normalized = pca.transform(test_data_normalized)

print "Choose best hyperparameters..."
gammas = np.logspace(0.001,0.005,20)
cs = np.logspace(1,4,20)

svc = svm.SVC(kernel='rbf')
clf = grid_search.GridSearchCV(estimator=svc, param_grid=[dict(gamma=gammas), dict(C=cs)],n_jobs=10)
clf.fit(train_data, target)

print "best hyperparameters:\nC:{0}\ngamma:{1}".format(clf.best_estimator_.C, clf.best_estimator_.gamma)

print "apply svm with best hyperparameters"
svc2 = svm.SVC(gamma=clf.best_estimator_.gamma, C=clf.best_estimator_.C)
svc2.fit(train_data_normalized, target)
test_y = svc2.predict(test_data_normalized)
pd.DataFrame({"ImageId": range(1,len(test_y)+1), "Label": map(int,test_y)}).to_csv('hyper_parameters_svm_normalized.csv', index=False, header=True)
```
